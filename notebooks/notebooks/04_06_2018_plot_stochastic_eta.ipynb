```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt

fontsize = 16;
X = np.load('2VL_results.npz');
tl_its = X['tl_its'];
se_its = X['se_its'];
K_etas = X['K_etas'];
plt.figure();
inds = np.array([0,1,3,4]);
plt.plot(K_etas, se_its[:,inds]);
plt.scatter(12*np.ones(4,), tl_its[inds], color='k');
plt.legend(['random seed 1', 'random seed 2', 'random seed 3', 'random seed 4']);
plt.xlabel('K_eta', fontsize=fontsize);
plt.ylabel('iterations to convergence', fontsize=fontsize);
plt.show();
```
