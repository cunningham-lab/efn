```python
def closest(support_point, Xgrid):
    #constructs an nxn grid and gives back the point on the grid which, after being transformed by the network
    #is closest to simplex_point
    diffs = Xgrid - support_point;
    dists = np.linalg.norm(diffs, ord=2, axis=1);
    clos_ind = np.argmin(dists);
    return clos_ind;

def gen_grid(sess, n, X, Z0, log_p_zs, eta, _eta, Tx_input, _Tx_input):
    K = _eta.shape[0];
    M = n*n;
    x = np.linspace(-3.5,3.5,n)
    _z0 = np.zeros((M,2));
    for i in range(n):
        for j in range(n):
            ind = n*i+j;
            _z0[ind,0] = x[i];
            _z0[ind,1] = x[j];
    z0 = np.expand_dims(np.expand_dims(_z0, 0), 3);
    feed_dict = {Z0:z0, eta:_eta, Tx_input:_Tx_input};
    Xgrid, log_p_z_grid = sess.run([X, log_p_zs], feed_dict);

    return Xgrid, log_p_z_grid, _z0;

def plotPoissonRatePosterior(Xgrid, log_p_z_grid, xlim, ylim, n_plot):
    x1 = np.linspace(0, xlim, n_plot)
    x2 = np.linspace(0, ylim, n_plot)
    sns.set_palette(sns.color_palette("Set1", n_colors=8, desat=.6))
    sns.set_style("whitegrid")
    sns.set_style("ticks")
    sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.0})
    sbpal = sns.color_palette()
    
    Z = np.zeros([n_plot,n_plot])
    for i in range(n_plot):
        for j in range(n_plot):
            support_point = np.array([[x1[i],x2[j]]])
            ind = closest(support_point, Xgrid);
            Z[i,j] = log_p_z_grid[ind];
    
    pvals = np.exp(Z);
    
    mycm = sns.light_palette("navy", as_cmap=True)
    mycm.set_under('w')
    plt.imshow(pvals.T, cmap=mycm, origin='lower')
    ax = plt.gca()
    ax.set_xlabel([0,xlim]);
    ax.set_xticklabels([]);
    ax.set_ylabel([0,ylim]);
    ax.set_yticklabels([]);
    plt.colorbar();
    return Z;

n_plot = 30;
Xgrid, log_p_z_grid, z0s = gen_grid(sess, n_plot, X, Z0, log_p_zs, eta, _eta, Tx_input, _Tx_input);

mu = params['mus'];
Sigma = params['Sigmas'];
xs = params['xs'];
x = xs[0];
xlim = 10;
ylim = 10;
plt.figure(figsize=(14, 6));
plt.subplot(1,3,1);
plotContourTruncatedNormal(mu[0], Sigma[0], xlim, ylim, 100, scatter=False);
plt.xlabel('z_1');
plt.ylabel('z_2');

plt.subplot(1,3,2);
mycm = sns.light_palette("navy", as_cmap=True)
mycm.set_under('w')
plot_range = [[0, xlim], [0, ylim]];
H, xedges, yedges = np.histogram2d(x[0], x[1], bins=8, range=plot_range, normed=True)
plt.imshow(H.T, origin='lower', interpolation='none', cmap=mycm);
plt.title('data sampled from Poisson([5,1]), N=10');
#plt.colorbar();


plt.subplot(1,3,3);
xlim = 10;
ylim = 10;
Z = plotPoissonRatePosterior(Xgrid[0,:,:,0], log_p_z_grid[0], xlim, ylim, n_plot);
plt.show();
```
