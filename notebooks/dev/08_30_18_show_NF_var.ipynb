```python
%matplotlib inline
from train_nf import train_nf
import numpy as np
from scipy.stats import dirichlet
from tf_util.families import family_from_str
from efn_util import model_opt_hps
import os, sys
from dirichlet import simplex
import matplotlib.pyplot as plt

os.chdir('../../');
```

```python
exp_fam = 'dirichlet';
D = 4;
nlayers = 2;
give_inverse_hint = False;
dist_seed = 0;
dir_str = 'bv_test';

TIF_flow_type, _, lr_order = model_opt_hps(exp_fam, D);

flow_dict = {'latent_dynamics':None, \
             'TIF_flow_type':TIF_flow_type, \
             'repeats':nlayers};

fam_class = family_from_str(exp_fam);
family = fam_class(D);

cost_type = 'KL';
M_eta = 1000;
min_iters = 20000;
max_iters = 20000;
check_rate = 100;

param_net_input_type = 'eta';  # I should generalize this draw_etas function to accept a None
np.random.seed(dist_seed);
eta, param_net_input, Tx_input, params = family.draw_etas(1, param_net_input_type, give_inverse_hint);
params = params[0];
params.update({'dist_seed':dist_seed});

alpha = params['alpha'];
print('alpha');
print(alpha);
dist = dirichlet(params['alpha']);
nsamps = 100;
X_true = dist.rvs(nsamps);
simplex.scatter(X_true)
plt.show()

```

```python
fontsize = 16;

for random_seed in range(10):
    log_p_zs, X, train_R2s, train_KLs, it = train_nf(family, params, flow_dict, cost_type, M_eta, lr_order, \
                                                     random_seed, min_iters, max_iters, check_rate, dir_str);
    plt.figure(figsize=(8,4));
    plt.subplot(1,2,1);
    simplex.scatter(X_true);
    plt.title('ground truth', fontsize=fontsize);
    plt.subplot(1,2,2);
    simplex.scatter(X[0,:100,:,0]);
    plt.title('trained NF', fontsize=fontsize);
    plt.show();
```
